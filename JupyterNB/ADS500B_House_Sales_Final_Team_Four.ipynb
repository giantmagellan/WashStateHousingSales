```python
#Find Missing Values of sqft_living. 

#Take Sum of sqft_above + sqft_basement.
```

```python
sqft_living_total = house_sales['sqft_above'] + house_sales['sqft_basement']
house_sales['sqft_living'] = house_sales['sqft_living'].fillna(sqft_living_total, axis=0)
```

```python
avglot_by_zipcode = house_sales.groupby('zipcode')['sqft_lot15'].mean()
```
